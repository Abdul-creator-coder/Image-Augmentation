```python
import albumentations as A
import cv2
from matplotlib import pyplot as plt
def visualize(image):
    plt.figure(figsize=(10, 10))
    plt.axis('off')
    plt.imshow(image)

image = cv2.imread("C:/Users/ghani/Downloads/Data Augmentation/Dog.jpg")
visualize(image) #Original Image

transform = A.Compose(
    [A.HorizontalFlip(always_apply=False, p=0.9),  #Horizontal flip
    A.VerticalFlip(always_apply=False, p=0.7),    #Vertical Flip
    A.Resize(always_apply=False, p=1.0, height=106, width=160, interpolation=0),#Resize the image
    A.RandomRotate90(),#Random rotation of image by 90 degree
    A.HueSaturationValue(),
    A.ToSepia(always_apply=False, p=1.0),#Applying sepia filter to image
    A.InvertImg(always_apply=False, p=0.3),#Inverting the image
    A.SmallestMaxSize(always_apply=False, p=1.0, max_size=363, interpolation=0)
    ])

augmented_image = transform(image=image)['image']
visualize(augmented_image)
```
